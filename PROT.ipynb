# PROT Translating RNA into Protein

The RNA codon table dictates the details regarding the encoding of specific codons into the amino acid alphabet.

**Given**: An RNA string s corresponding to a strand of mRNA (of length at most 10 kbp).

**Return**: The protein string encoded by s


```python
f = open("codons.txt") #copy-pasted from Rosalind into txt file
dict={}
for line in f:
    temp=line.strip() #takes current line and puts it into temp string
    temp2=[x.strip() for x in temp.split()] #split the temp string into a list, remove white spaces
    #print(temp2)
    for i in range (len(temp2)-1):
        if(i%2==0):
            dict[temp2[i]]=temp2[i+1]
```

```python
input = open("rosalind_prot.txt")
for line in input:
    rna=line.strip()
answer=[]

for i in range(len(rna)-5):
    if(i%3==0):
        answer.append(dict[rna[i:i+3]])

print(''.join(answer))
```
